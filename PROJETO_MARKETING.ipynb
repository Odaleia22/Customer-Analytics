```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo CSV
df = pd.read_csv("/content/drive/MyDrive/Marketing/Marketing-Tratado.csv")

# Remover os estados civis "Absurdo" e "YOLO"
df = df[~df["Estado_Civil"].isin(["Absurdo", "YOLO"])]

# Agrupar os clientes por idade e calcular a média da renda para cada grupo
renda_media_por_idade = df.groupby("Idade")["Renda"].mean()

# Criar o gráfico de dispersão
plt.scatter(renda_media_por_idade.index, renda_media_por_idade.values)

# Configurar os títulos do gráfico e dos eixos
plt.title("Renda Média por Idade")
plt.xlabel("Idade")
plt.ylabel("Renda Média")

# Mostrar o gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregando o dataset
df = pd.read_csv('/content/drive/MyDrive/Marketing/Marketing-Tratado.csv')

# Calculando a porcentagem de cada nível educacional
educacao_porcentagem = df['Educação'].value_counts(normalize=True) * 100

# Criando cores em tons de azul
cores = plt.cm.Blues(educacao_porcentagem.values / max(educacao_porcentagem.values))

# Criando o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(educacao_porcentagem, labels=educacao_porcentagem.index, autopct='%1.1f%%', startangle=90, colors=cores)

# Adicionando título
plt.title('Níveis educacionais')

# Ajustando a aparência do gráfico
plt.axis('equal')  # Garante que o gráfico seja um círculo
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Carregar os dados do arquivo CSV
df = pd.read_csv("/content/drive/MyDrive/Marketing/Marketing-Tratado.csv")

# Remover os estados civis "Absurdo" e "YOLO"
df = df[~df["Estado_Civil"].isin(["Absurdo", "YOLO"])]

# Agrupar os clientes por estado civil e contar a frequência de cada grupo
estado_civil_counts = df["Estado_Civil"].value_counts()

# Criar um gradiente de cores azuis
cmap = mcolors.LinearSegmentedColormap.from_list("my_cmap", ["#0000FF", "#0080FF"])

# Criar o gráfico de colunas com o gradiente de cores
plt.bar(
    estado_civil_counts.index,
    estado_civil_counts.values,
    color=cmap(estado_civil_counts.values / estado_civil_counts.values.max()),
)

# Configurar os títulos do gráfico e dos eixos
plt.title("Quantidade de Clientes por Estado Civil")
plt.xlabel("Estado Civil")
plt.ylabel("Quantidade de Clientes")

# Ajustar o layout do gráfico para evitar sobreposição de labels
plt.xticks(rotation=45, ha="right")

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
